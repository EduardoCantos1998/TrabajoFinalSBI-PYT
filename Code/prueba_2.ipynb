```python
import math

nan_list = []
for i in df_dict.keys():
    for j in df_dict[i]:
        for k in df_dict[i][j]:
            if math.isnan(k):
                nan_list.append(i)
print(len(set(nan_list)))
```

```python
for i in df_dict.keys():
    for j in df_dict[i]:
        for index, k in enumerate(df_dict[i][j]):
            if math.isnan(k):
                df_dict[i][j][index] = 7
```

```python
amino_keys = {
    'ALA': 0, 'ARG': 1, 'ASN': 2, 'ASP': 3, 'CYS': 4, 'GLN': 5, 'GLU': 6, 'GLY': 7, 
    'HIS': 8, 'ILE': 9, 'LEU': 10, 'LYS': 11, 'MET': 12, 'PHE': 13, 'PRO': 14, 
    'SER': 15, 'THR': 16, 'TRP': 17, 'TYR': 18, 'VAL': 19
    }
SS_dict = {'-': 0, 'B': 1, 'T': 2, 'S': 3, 'G': 4, 'E': 5, 'H': 6, 'I': 7}
    
def flipper(X):
    new_X = {}
    for i in X.keys():
        new_X[X[i]] = i
    return new_X 

new_amino_keys = flipper(amino_keys)
new_SS = flipper(SS_dict)

print(new_amino_keys)
print(new_SS)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Split the data into training, validation and test sets
class_weights = {0: 1, 1: 7}
model = RandomForestClassifier(max_depth=2,random_state=0, class_weight=class_weights)

print("Spliting the keys in train/test/validation sets.")
train_keys, test_keys = train_test_split(list(df_dict.keys()), test_size=0.9, random_state=0)
train_keys, val_keys = train_test_split(train_keys, test_size=0.2, random_state=0)
```

```python
X_train = pd.concat([df_dict[key].drop("BINDING_ATOM", axis=1) for key in train_keys])
y_train = pd.concat([df_dict[key].BINDING_ATOM for key in train_keys])
for i in train_keys:
    model.fit(X_train, y_train)
```

```python
accuracy = []
for i in test_keys:
    acc = accuracy_score(model.predict(df_dict[i].drop("BINDING_ATOM", axis = 1)), df_dict[i]["BINDING_ATOM"])
    print(acc)
    accuracy.append(acc)
print(f"Average accuracy: {sum(accuracy)/len(accuracy)}")
```
